```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from dateutil.parser import parse
```

```python
f = '../data/sismos/SSNMX_catalogo_19000101_20241011_GRO.csv'
sismos = pd.read_csv(f,skiprows=4,skipfooter=7, engine='python')
sismos['time'] = sismos.Fecha + ' ' + sismos.Hora
sismos.Magnitud =  sismos.Magnitud.str.replace('no calculable','nan')
sismos.Magnitud = sismos.Magnitud.astype(float)
sismos['time'] = pd.to_datetime(sismos.time,format='%Y-%m-%d %H:%M:%S')
sismos['Mayor5'] = sismos['Magnitud'].apply(lambda x: x if x >= 5 else None)
sismos['Menor5'] = sismos['Magnitud'].apply(lambda x: x if x < 5 else None)

sismos=sismos.loc[sismos.time.dt.year >= 1980]
sismos.head()
```

```python
sismos5mayores = sismos[(sismos.Magnitud >= 5) & (sismos.time.dt.year >= 1980)] .copy()
sismos5mayores
```

```python
len(sismos.Mayor5.dropna()),len(sismos.Menor5.dropna())
```

```python
fig, (ax_hist, ax) = plt.subplots(2, 1, figsize=(10, 6), gridspec_kw={'height_ratios': [1, 1]})

# Histogram plot
sismos.Magnitud.hist(ax=ax_hist, bins=30)

# Set x-ticks every 0.5
ax_hist.set_xticks([i * 0.5 for i in range(int(ax_hist.get_xlim()[1] / 0.5) + 1)])

# Remove the top and right spines for the histogram
ax_hist.spines['top'].set_visible(False)
ax_hist.spines['right'].set_visible(False)

# Remove the grid
ax_hist.grid(False)

# Set tick labels to gray color
ax_hist.tick_params(axis='x', colors='gray')
ax_hist.tick_params(axis='y', colors='gray')

# Set y-axis labels with thousands separator
ax_hist.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Set y-axis label
ax_hist.set_ylabel("Número de sismos", color='gray')

# Scatter plot
f1 = parse("1980-01-01")
f2 = parse("2024-12-1")
ax.scatter(sismos.time, sismos.Mayor5)
ax.scatter(sismos.time, sismos.Menor5, c='gray')

ax.set_ylim(0, 9)
ax.set_xlim(f1, f2)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Remove the grid
ax.grid(False)

# Set tick labels to gray color
ax.tick_params(axis='x', colors='gray')
ax.tick_params(axis='y', colors='gray')

# Set y-axis label
ax.set_ylabel("Magnitud", color='gray')

# Remove the x-axis label
ax.set_xlabel("")

plt.tight_layout()
plt.show()
```

```python
# df = sismos
# # Crear el DataFrame (ya debes tener los datos cargados en df)
# # df = pd.read_csv('ruta_a_tu_archivo.csv')

# # Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([-103, -95, 15, 21], crs=ccrs.PlateCarree())  # Coordenadas de la región de Guerrero

# Agregar características del mapa
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle='-')

# Graficar los sismos
sc = ax.scatter(
    sismos5mayores['Longitud'], sismos5mayores['Latitud'], 
    c=sismos5mayores['Magnitud'], s=sismos5mayores['Magnitud']**3,  # Ajustar tamaño y color según magnitud
    cmap='Reds', alpha=0.6, edgecolors='k', 
    transform=ccrs.PlateCarree(), 
)

# Agregar una barra de color para la magnitud
cbar = plt.colorbar(sc, ax=ax, orientation='vertical', shrink=0.5, pad=0.05)
cbar.set_label('Magnitud')

# Agregar título y etiquetas
ax.set_title('Sismos mayores a 5 en Guerrero')
# plt.legend()

# Mostrar y/o guardar la figura
# plt.savefig('sismos_guerrero.png', dpi=300, bbox_inches='tight')
plt.show()
len(sismos5mayores)
```
