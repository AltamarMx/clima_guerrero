```python
import tropycal.tracks as tracks
import matplotlib.pyplot as plt
import numpy as np
```

```python
basin = tracks.TrackDataset(basin='east_pacific',source='hurdat',include_btk=False)
```

```python
season2017 = basin.get_season(2017)
season2018 = basin.get_season(2018)
season2019 = basin.get_season(2019)
season2020 = basin.get_season(2020)
season2021 = basin.get_season(2021)
season2022 = basin.get_season(2022)
season2023 = basin.get_season(2023)
# season2024 = basin.get_season(2024)

combined = season2017 + season2018 + season2019 + season2020 + season2021 + season2022 + season2023
```

```python
combined.get_storm_id
```

```python
combined.plot(    
    prop={'dots': False, 'linecolor': 'category', 'linewidth': 1.0, 'plot_names': False, 'plot_grid': False},
    map_prop={
        'plot_gridlines': False,
        'res': 'h',                      # Alta resolución de fronteras
        'linewidth': 1.0,                # Grosor de las líneas
        'linecolor': 'black',            # Color de los límites políticos
        'state_alpha': 0.1               # Opacidad de los límites de estados
    },
    figsize=(10, 4)  # Establecer el tamaño de la figura
)
```

```python

# Datos proporcionados
years = ['2017', '2018', '2019', '2020', '2021', '2022', '2023']
total_storms = [20, 26, 21, 21, 19, 19, 20]
hurricanes = [9, 13, 7, 4, 8, 10, 10]
major_hurricanes = [4, 10, 4, 3, 2, 4, 8]
season_ace = [100.8, 317.9, 97.7, 77.2, 94.5, 114.6, 165.3]

# Gráfica de barras acumuladas para tormentas totales, huracanes y huracanes mayores
x = np.arange(len(years))  # Rango para las etiquetas del eje X
width = 0.4  # Ancho de las barras más angostas

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(x, major_hurricanes, width, label='Huracanes mayores', color='#4c72b0')
ax.bar(x, hurricanes, width, bottom=major_hurricanes, label='Huracanes', color='#55a868')
ax.bar(x, total_storms, width, bottom=np.array(hurricanes) + np.array(major_hurricanes), label='Tormentas', color='#c44e52')

# Añadir el número de eventos para cada categoría dentro de cada barra
for i in range(len(years)):
    ax.text(x[i], major_hurricanes[i] / 2, str(major_hurricanes[i]), color='white', ha='center', va='center')
    ax.text(x[i], major_hurricanes[i] + hurricanes[i] / 2, str(hurricanes[i]), color='white', ha='center', va='center')
    ax.text(x[i], major_hurricanes[i] + hurricanes[i] + total_storms[i] / 2, str(total_storms[i]), color='white', ha='center', va='center')

# Ajustes de estilo
theme_color = 'grey'
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(theme_color)
ax.spines['bottom'].set_color(theme_color)
ax.tick_params(colors=theme_color)
ax.set_ylabel('Número de eventos', color=theme_color)
ax.set_xticks(x)
ax.set_xticklabels(years, color=theme_color)
ax.legend(frameon=False)

plt.tight_layout()
plt.show()
```
