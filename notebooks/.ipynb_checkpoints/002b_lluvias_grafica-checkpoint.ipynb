```python
lluvia = pd.read_parquet("../data/lluvia/lluvia.parquet")
estaciones = lluvia.ESTACION.unique()
lluvia.head(3)
```

```python
top = lluvia[lluvia.index.year==2024].sort_values(by="lluvia",ascending=False)["ESTACION"].head(1).to_list()
lluvia = lluvia[lluvia.ESTACION.isin(top)]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Asegúrate de que el índice sea de tipo datetime
lluvia.index = pd.to_datetime(lluvia.index)

fig, ax = plt.subplots(figsize=(12, 5))

# Graficar cada estación y añadir leyenda al final de cada línea
for estacion in top:
    datos_estacion = lluvia[lluvia['ESTACION'] == estacion]
    ax.plot(datos_estacion.index, datos_estacion['lluvia'], label=estacion, linewidth=2)

    # Obtener el último valor del índice y la última coordenada y
    ultimo_punto_x = datos_estacion.index[-1]
    ultimo_punto_y = datos_estacion['lluvia'].iloc[-1]

    # Añadir el nombre de la estación al final de la línea
    # ax.text(ultimo_punto_x + pd.Timedelta('5D'), ultimo_punto_y, estacion, fontsize=10, ha='left', va='center', color='gray')

# Encontrar y marcar el valor máximo de cada año
lluvia_anual_max = lluvia.resample('YE')['lluvia'].max()

for year, max_value in lluvia_anual_max.items():
    if pd.notna(max_value):
        max_date = lluvia[lluvia['lluvia'] == max_value].index[0]
        ax.text(max_date, max_value, f'{max_value:.1f} mm', fontsize=10, ha='right', va='bottom', color='k',alpha=0.8)

# Añadir línea vertical y anotación para Huracán Otis
fecha_otis = pd.Timestamp('2023-10-01')
ax.axvline(x=fecha_otis, color='grey', linestyle='-', linewidth=0.5, alpha=1)
ax.text(fecha_otis + pd.Timedelta("105D"), 1200, 'Huracán Otis', ha='right', color='black', alpha=0.8)

# Añadir línea vertical y anotación para Huracán John
fecha_john = pd.Timestamp('2024-09-01')
ax.axvline(x=fecha_john, color='grey', linestyle='-', linewidth=0.5, alpha=1)
ax.text(fecha_john + pd.Timedelta("15D"), 1200, 'Huracán John', ha='left', color='black', alpha=0.8)

# Configuración adicional para mejorar la narrativa
ax.spines[['top', 'right']].set_visible(False)

# Cambiar el color de los spines y ticks a gris claro
for spine in ax.spines.values():
    spine.set_color('gray')

ax.tick_params(colors='gray')
ax.yaxis.label.set_color('gray')
ax.xaxis.label.set_color('gray')

# Cambiar el color de los textos de los ticks
ax.xaxis.set_tick_params(labelcolor='gray')
ax.yaxis.set_tick_params(labelcolor='gray')

# Cambiar el color de la etiqueta del eje y
ax.set_ylabel("Lluvia [mm]", fontsize=12, color='gray')

# Mostrar la gráfica
plt.show()
```
