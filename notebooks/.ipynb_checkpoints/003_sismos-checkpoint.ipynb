```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
f = '../data/sismos/SSNMX_catalogo_19000101_20241011_GRO.csv'
sismos = pd.read_csv(f,skiprows=4,skipfooter=7, engine='python')
sismos['time'] = sismos.Fecha + ' ' + sismos.Hora
sismos.Magnitud =  sismos.Magnitud.str.replace('no calculable','nan')
sismos.Magnitud = sismos.Magnitud.astype(float)
sismos['time'] = pd.to_datetime(sismos.time,format='%Y-%m-%d %H:%M:%S')
sismos['Mayor5'] = sismos['Magnitud'].apply(lambda x: x if x >= 5 else None)
sismos['Menor5'] = sismos['Magnitud'].apply(lambda x: x if x < 5 else None)

sismos.head()
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.scatter(sismos.time,sismos.Mayor5)
ax.scatter(sismos.time,sismos.Menor5, c ='gray')

ax.set_ylim(0,9)
ax.grid()
```

```python

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([-103, -95, 15, 21], crs=ccrs.PlateCarree())  # Coordenadas de la regin de Guerrero

# Agregar caractersticas del mapa
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle='-')

# Graficar los sismos
ax.scatter(
    sismos['Longitud'], sismos['Latitud'],  # Ajustar tamao y color segn magnitud
    alpha=0.6, edgecolors='k', 
    transform=ccrs.PlateCarree(), label="Sismos"
)

# Agregar una barra de color para la magnitud
# cbar = plt.colorbar(sc, ax=ax, orientation='vertical', shrink=0.5, pad=0.05)
# cbar.set_label('Magnitud')

# Agregar ttulo y etiquetas
ax.set_title('Sismos en Guerrero, Mxico desde 1900')
plt.legend()

# Mostrar y/o guardar la figura
plt.savefig('sismos_guerrero.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sismos5mayores = sismos[(sismos.Magnitud >= 5) & (sismos.time.dt.year >= 1970)] .copy()
sismos5mayores
```

```python
# df = sismos
# # Crear el DataFrame (ya debes tener los datos cargados en df)
# # df = pd.read_csv('ruta_a_tu_archivo.csv')

# # Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([-103, -95, 15, 21], crs=ccrs.PlateCarree())  # Coordenadas de la regin de Guerrero

# Agregar caractersticas del mapa
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle='-')

# Graficar los sismos
sc = ax.scatter(
    sismos5mayores['Longitud'], sismos5mayores['Latitud'], 
    c=sismos5mayores['Magnitud'], s=sismos5mayores['Magnitud']**3,  # Ajustar tamao y color segn magnitud
    cmap='Reds', alpha=0.6, edgecolors='k', 
    transform=ccrs.PlateCarree(), label="Sismos"
)

# Agregar una barra de color para la magnitud
cbar = plt.colorbar(sc, ax=ax, orientation='vertical', shrink=0.5, pad=0.05)
cbar.set_label('Magnitud')

# Agregar ttulo y etiquetas
ax.set_title('Sismos mayores a 5 en Guerrero, Mxico desde 1970')
plt.legend()

# Mostrar y/o guardar la figura
# plt.savefig('sismos_guerrero.png', dpi=300, bbox_inches='tight')
plt.show()
len(sismos5mayores)
```
